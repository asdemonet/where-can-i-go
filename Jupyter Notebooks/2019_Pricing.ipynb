```python
import pandas as pd
import numpy as np
```

```python
coupon_2019_q1_df=pd.read_csv("Q1_COUPON.csv")
coupon_2019_q2_df=pd.read_csv("Q2_COUPON.csv")
coupon_2019_q3_df=pd.read_csv("Q3_COUPON.csv")
coupon_2019_q4_df=pd.read_csv("Q4_COUPON.csv")
ticket_2019_q1_df=pd.read_csv("Q1_TICKET.csv")
ticket_2019_q2_df=pd.read_csv("Q2_TICKET.csv")
ticket_2019_q3_df=pd.read_csv("Q3_TICKET.csv")
ticket_2019_q4_df=pd.read_csv("Q4_TICKET.csv")
```

```python
COUPON_2019_Q1_RDU_df = coupon_2019_q1_df.loc[(coupon_2019_q1_df["ORIGIN"] == "RDU"),:]
COUPON_2019_Q2_RDU_df = coupon_2019_q2_df.loc[(coupon_2019_q2_df["ORIGIN"] == "RDU"),:]
COUPON_2019_Q3_RDU_df = coupon_2019_q3_df.loc[(coupon_2019_q3_df["ORIGIN"] == "RDU"),:]
COUPON_2019_Q4_RDU_df = coupon_2019_q4_df.loc[(coupon_2019_q4_df["ORIGIN"] == "RDU"),:]
COUPON_2019_Q1_RDU_df.head()
```

```python
COUPON_2019_Q1_NONSTOP_df = COUPON_2019_Q1_RDU_df.loc[(COUPON_2019_Q1_RDU_df["COUPONS"] == 2),:]
COUPON_2019_Q2_NONSTOP_df = COUPON_2019_Q2_RDU_df.loc[(COUPON_2019_Q2_RDU_df["COUPONS"] == 2),:]
COUPON_2019_Q3_NONSTOP_df = COUPON_2019_Q3_RDU_df.loc[(COUPON_2019_Q3_RDU_df["COUPONS"] == 2),:]
COUPON_2019_Q4_NONSTOP_df = COUPON_2019_Q4_RDU_df.loc[(COUPON_2019_Q4_RDU_df["COUPONS"] == 2),:]
COUPON_2019_Q4_NONSTOP_df.head()
```

```python
TICKET_2019_Q1_ROUNDTRIP_df = ticket_2019_q1_df.loc[(ticket_2019_q1_df["ROUNDTRIP"] == 1),:]
TICKET_2019_Q2_ROUNDTRIP_df = ticket_2019_q2_df.loc[(ticket_2019_q2_df["ROUNDTRIP"] == 1),:]
TICKET_2019_Q3_ROUNDTRIP_df = ticket_2019_q3_df.loc[(ticket_2019_q3_df["ROUNDTRIP"] == 1),:]
TICKET_2019_Q4_ROUNDTRIP_df = ticket_2019_q4_df.loc[(ticket_2019_q4_df["ROUNDTRIP"] == 1),:]
```

```python
MERGED_Q1=pd.merge(COUPON_2019_Q1_NONSTOP_df, TICKET_2019_Q1_ROUNDTRIP_df, on=["ITIN_ID"])
MERGED_Q2=pd.merge(COUPON_2019_Q2_NONSTOP_df, TICKET_2019_Q2_ROUNDTRIP_df, on=["ITIN_ID"])
MERGED_Q3=pd.merge(COUPON_2019_Q3_NONSTOP_df, TICKET_2019_Q3_ROUNDTRIP_df, on=["ITIN_ID"])
MERGED_Q4=pd.merge(COUPON_2019_Q4_NONSTOP_df, TICKET_2019_Q4_ROUNDTRIP_df, on=["ITIN_ID"])
```

```python
MERGE_2019_df = MERGED_Q1.append(MERGED_Q2)
MERGE_2019_df = MERGE_2019_df.append(MERGED_Q3)
MERGE_2019_df = MERGE_2019_df.append(MERGED_Q4)
```

```python
MERGED_2019=MERGE_2019_df.drop('Unnamed: 14',axis=1)
MERGED_2019_FINAL=MERGED_2019.drop('Unnamed: 4',axis=1)
MERGED_2019_FINAL
```

```python
MERGED_2019_FINAL['YEAR']='2019'
MERGED_2019_FINAL
```

```python
MERGED_2019_FINAL=MERGED_2019_FINAL[["ITIN_ID","COUPONS","YEAR","ORIGIN_AIRPORT_ID","QUARTER","ORIGIN","DEST_AIRPORT_ID","DEST","TICKET_CARRIER","OPERATING_CARRIER","REPORTING_CARRIER","PASSENGERS","FARE_CLASS","DISTANCE_GROUP","ITIN_GEO_TYPE","ROUNDTRIP","ITIN_FARE","MILES_FLOWN"]]
MERGED_2019_FINAL
```

```python
MERGED_2019_FINAL.to_csv('MERGE_2019_df.csv')
```
