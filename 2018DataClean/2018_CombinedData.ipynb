```python
# Read in Pandas

import pandas as pd
```

```python
# Read in Cleaned Quarterly CSVs to Dataframes

Q12018_df = pd.read_csv('CleanQuarterCSVs/Combined2018Q1')
Q22018_df = pd.read_csv('CleanQuarterCSVs/Combined2018Q2')
Q32018_df = pd.read_csv('CleanQuarterCSVs/Combined2018Q3')
Q42018_df = pd.read_csv('CleanQuarterCSVs/Combined2018Q4')
```

```python
# Created 2018 Dataframe. Use .append() to add each quarter

YR2018_df = Q12018_df.append(Q22018_df).append(Q32018_df).append(Q42018_df)

# Display data for the year of 2018

YR2018_df
```

```python
YR2018_df.to_csv('YR2018_Final_CSV/YR2018Data.csv',index=False)
```
